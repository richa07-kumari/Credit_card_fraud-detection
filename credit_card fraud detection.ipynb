Importing relevant libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset to a Pandas Dataframe
credit_card=pd.read_csv('Downloads/creditcard.csv')
```

```python
# To extract first 5 rows
credit_card.head()
```

```python
# To extract last 5 rows
credit_card.tail()
```

```python
credit_card.info()
```

```python
credit_card.isnull().sum()
```

```python
# Distribution of genuine data and fraud data
credit_card['Class'].value_counts()
```

```python
# Separating the data for analysis
legit=credit_card[credit_card['Class']==0]
fraud=credit_card[credit_card.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for both the transactions
credit_card.groupby('Class').mean() # groupby the two classes
```

We will perform undersampling

Build a sample dataset containing similar distribution of normal and fraud transaction

Number of fraud transaction --> 492

So we will take randon 492 genuine transaction data and then build model

```python
legit_sample=legit.sample(n=492)
```

Concatenating two DataFrames

```python
new_dataset=pd.concat([legit_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the dataset into features and targets

```python
X=new_dataset.drop(columns='Class', axis=1)
Y=new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

#Split the data into train data and test data

```python
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression, as we have binary predictors

```python
model=LogisticRegression()
```

```python
model.fit(X_train, Y_train)
```

Model evaluation(based on accuracy score)

```python
#accuracy on training data
X_train_prediction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_prediction, Y_train)
```

```python
print("Training data accuracy:", training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction, Y_test)
```

```python
print("Test data accuracy:", test_data_accuracy)
```
